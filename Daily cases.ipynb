```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sars_data = pd.read_csv("sars_2003_complete_dataset_clean.csv")
sars_data
```

```python
sars_data.head()
```

```python
keep_columns = ["Date","Cumulative number of case(s)","Number of deaths","Number recovered"]

sars_data1 = sars_data[keep_columns]
sars_data1
```

```python
sars_data2 = sars_data1.sort_values("Date",ascending = True)
sars_data2
```

```python
sars_data3 = sars_data2.groupby(sars_data2["Date"]).sum()
sars_data3
```

```python
sars_data3["Daily case(s)"] = sars_data3["Cumulative number of case(s)"].diff()
sars_data3
```

```python
sars_data4 = sars_data3.loc[sars_data3["Daily case(s)"]>=0]
sars_data4
```

```python
sars_data3["Daily case(s)"].plot(kind="line")
```
